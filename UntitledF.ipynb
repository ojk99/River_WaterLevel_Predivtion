```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
import tensorflow as tf
import glob

######################################################################

# TensorFlow에서 GPU를 사용할 수 있는지 확인
print("GPU 사용 가능 여부:", tf.config.list_physical_devices('GPU'))
# TensorFlow 버전 확인
print("TensorFlow 버전:", tf.__version__)

######################################################################

# CSV 파일 불러오기
def rivermake(file_years):

    # 빈 데이터프레임 생성
    river_concat = pd.DataFrame()

    # 반복문을 통해 파일을 읽어들이고 연도에 따라 데이터프레임에 추가
    for year in file_years:
        file_path = f"C:\\yolo\\tf\\csv\\여수대교{year}.csv"
        df = pd.read_csv(file_path)
        df['time'] = pd.to_datetime(df['시간'])
        df.rename(columns={'수위': 'level'}, inplace=True)
        df.drop(['관측소명'], axis=1, inplace=True)
        river_concat = pd.concat([river_concat, df])
    return river_concat

# 필요한 변수 설정
river = rivermake(("2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014", "2015", "2016", "2017", "2018", "2019", "2020")).copy()
river2 = rivermake(("2009","2010","2011","2012","2013","2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021",)).copy()

# 파일 패턴 지정
file_pattern = "C:\\yolo\\tf\\csv\\여수대교*.csv"

# 파일들의 경로를 검색하여 리스트로 반환
file_paths = glob.glob(file_pattern)

# 파일들을 순회하며 읽어들이기
dataframes = [pd.read_csv(file_path) for file_path in file_paths]

# 파일들을 하나로 합치기
river3 = pd.concat(dataframes)

# river3 필요없는 열 삭제
river3.drop(['관측소명'], axis=1, inplace=True)
# river3 '수위' 열 이름 변경
river3.rename(columns={'수위': 'level'}, inplace=True)
# river3 '시간' 열을 datetime 형식으로 변환
river3['time'] = pd.to_datetime(river3['시간'])

a8 = pd.read_csv("C:\\yolo\\tf\\csv\\여수대교2021.csv")
# a8필요없는 열 삭제
a8.drop(['관측소명'], axis=1, inplace=True)
# a8 '시간' 열을 datetime 형식으로 변환
a8['time'] = pd.to_datetime(a8['시간'])
# a8 '수위' 열 이름 변경
a8.rename(columns={'수위': 'level'}, inplace=True)

######################################################################

# 모델 구성 및 학습
# 훈련 데이터 설정
train_data = river['level'].values
# 테스트 데이터 설정
test_data = a8['level'].values

# 훈련 데이터의 시간 열
train_dates = pd.to_datetime(river['time'], format='%Y-%m-%D')
# 테스트 데이터의 시간 열
test_dates = pd.to_datetime(a8['time'], format='%Y-%m-%D')

# 데이터 정규화
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data.reshape(-1, 1))

# LSTM을 위한 데이터 재구성
pred_days = 7  # 예측 기간
seq_len = 30  # 시퀀스 길이 = 미래 예측을 위한 과거 일 수

# LSTM 모델에 사용될 훈련 데이터 생성
trainX = []# LSTM 모델의 입력 데이터
trainY = []# LSTM 모델의 출력 데이터

for i in range(seq_len, len(train_data_scaled) - pred_days + 1):
    trainX.append(train_data_scaled[i - seq_len:i, 0])# trainX에 시퀀스 데이터 추가
    trainY.append(train_data_scaled[i + pred_days - 1:i + pred_days, 0])# trainY에 해당 시퀀스 다음 날짜의 값을 추가

trainX, trainY = np.array(trainX), np.array(trainY)# NumPy 배열로 변환
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))# LSTM 모델에 입력될 형태로 reshape

# LSTM 모델 구성
model = Sequential()
model.add(LSTM(64, input_shape=(trainX.shape[1], 1), return_sequences=True))
model.add(LSTM(32, return_sequences=False))
model.add(Dense(1))

# 모델 컴파일
#model.compile(optimizer='adam', loss='mse')
# 모델 컴파일
model.compile(optimizer='adam',
            loss = 'mse',
            metrics = ['mae'])

# GPU를 사용하여 행렬 곱셈 연산 수행
if tf.config.list_physical_devices('GPU'):
    print("GPU 사용")
    # GPU가 사용 가능한 경우
    with tf.device('/GPU:0'):
        history = model.fit(trainX, trainY, epochs=1, batch_size=32, validation_split=0.1, verbose=1)
else:
    print("GPU 미사용")
    # GPU가 사용할 수 없는 경우 또는 TensorFlow 버전이 GPU를 지원하지 않는 경우
    history = model.fit(trainX, trainY, epochs=1, batch_size=32, validation_split=0.1, verbose=1)
    
# 모델 저장
model.save("./my_model.h5")

######################################################################

# 테스트 데이터, 테스트 데이터를 예측한 데이터
# 테스트 데이터의 정규화
test_data_scaled = scaler.transform(test_data.reshape(-1, 1))

# LSTM 모델에 사용될 테스트 데이터 생성
testX = []# LSTM 모델의 입력 데이터
testY = []# LSTM 모델의 출력 데이터

# 테스트 데이터 생성 반복문
for i in range(seq_len, len(test_data_scaled) - pred_days + 1):
    testX.append(test_data_scaled[i - seq_len:i, 0])# testX에 시퀀스 데이터 추가
    testY.append(test_data_scaled[i + pred_days - 1:i + pred_days, 0])# testY에 해당 시퀀스 다음 날짜의 값을 추가

testX, testY = np.array(testX), np.array(testY)# NumPy 배열로 변환
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))# LSTM 모델에 입력될 형태로 reshape

# 예측 수행
prediction = model.predict(testX)

# 역변환
prediction = scaler.inverse_transform(prediction)

# 그래프 그리기
plt.figure(figsize=(14, 10))
plt.plot(test_dates[seq_len:seq_len+len(prediction)], test_data[seq_len:seq_len+len(prediction)], color='blue', label='Actual Level Price')
plt.plot(test_dates[seq_len:seq_len+len(prediction)], prediction, color='red', linestyle='--', label='Predicted Level Price')
plt.xlabel('Date')
plt.ylabel('Level Price(m)')
plt.title('Actual and Predicted Level Price')
plt.legend()
plt.show()

######################################################################

# 전체 데이터 중 일부분 추출
# 시작 날짜와 종료 날짜 설정
start_date = pd.Timestamp("2021-01-01")
end_date = pd.Timestamp("2022-01-01")

# 시작 날짜와 종료 날짜 사이의 데이터 필터링
filtered_river3 = river3[(river3['time'] >= start_date) & (river3['time'] <= end_date)]

# 필터링된 데이터에서 시간 열과 레벨 데이터 추출
filtered_dates = filtered_river3['time']
filtered_levels = filtered_river3['level']

# 그래프 그리기
plt.figure(figsize=(14, 10))
plt.plot(filtered_dates, filtered_levels, color='green', label='Filtered Data')
plt.xlabel('Date')
plt.ylabel('Level Price(m)')
plt.title('Filtered River Level Data')
plt.legend()
plt.show()

######################################################################

# 위험 수위를 초과하는 예측 및 해당 날짜 표시
# 마지막 유효한 인덱스 찾기
last_valid_index = train_dates.last_valid_index()
last_valid_date = train_dates[last_valid_index]

# NaN 값 제거 후 마지막 값 선택
last_valid_date = train_dates.dropna().iloc[-1]

# 훈련 데이터의 마지막 날짜 확인 #19910   2020-10-15 23:50:00
print("훈련 데이터의 마지막 날짜:", last_valid_date)

# 우이천 위험수위: 32.81
# 탄천 위험수위: 22.67

# 위험 수위를 초과하는 예측 및 해당 날짜 표시
threshold = 15.9 # 위험수위, 현제는 확인을 위한 임시 값
print(f"현제 설정된 위험수위 : {threshold}")
dangerous_predictions = []
for i in range(len(prediction)):
    if prediction[i] >= threshold:
        dangerous_predictions.append((prediction[i], test_dates[seq_len + i]))

if len(dangerous_predictions) > 0:
    print("위험 수위를 초과하는 예측:")
    for pred, date in dangerous_predictions:
        print("예측:", pred, "날짜:", date)
else:
    print("위험 수위를 초과하는 예측이 없습니다.")

######################################################################

# 훈련데이터, 테스트 데이터, 테스트 데이터를 예측한 데이터
# 그래프 그리기
plt.figure(figsize=(14, 10))
plt.plot(train_dates, train_data, color='black', label='Training Data')
plt.plot(test_dates[seq_len:seq_len+len(prediction)], test_data[seq_len:seq_len+len(prediction)], color='blue', label='Actual Test Data')
plt.plot(test_dates[seq_len:seq_len+len(prediction)], prediction, color='red', linestyle='--', label='Predicted Test Data')
#plt.plot(river2_time, river2_prediction, color='green', linestyle='-.', label='Predicted Data for Entire Range')
plt.xlabel('Date')
plt.ylabel('Level Price(m)')
plt.title('Actual and Predicted Level Price')
plt.legend()
plt.show()

######################################################################

# 데이터에 없는 새로운 날짜 범위의 예측 수위 그래프
# 새로운 날짜 범위 생성 (10분 간격)
new_dates = pd.date_range(start="2021-07-22 00:00", end="2021-10-15 23:50", freq='10T')
print(new_dates)

# 테스트 데이터의 정규화
test_data_scaled2 = scaler.transform(train_data.reshape(-1, 1))

# LSTM 모델에 사용될 테스트 데이터 생성
testX2 = []# LSTM 모델의 입력 데이터
testY2 = []# LSTM 모델의 출력 데이터

# 테스트 데이터 생성 반복문
for i in range(seq_len, len(test_data_scaled2) - pred_days + 1):
    testX2.append(test_data_scaled2[i - seq_len:i, 0])# testX에 시퀀스 데이터 추가
    testY2.append(test_data_scaled2[i + pred_days - 1:i + pred_days, 0])# testY에 해당 시퀀스 다음 날짜의 값을 추가

testX2, testY2 = np.array(testX2), np.array(testY2)# NumPy 배열로 변환
testX2 = np.reshape(testX2, (testX2.shape[0], testX2.shape[1], 1))# LSTM 모델에 입력될 형태로 reshape

# 예측 수행
prediction2 = model.predict(testX2)

# 역변환
prediction2 = scaler.inverse_transform(prediction2)

# 예측 결과(prediction)을 1차원 배열로 변환
prediction_1d = prediction2.flatten()
# 예측 결과(prediction)을 1차원 배열로 변환
prediction_1d = prediction2[:len(new_dates[seq_len:seq_len+len(prediction2)])].flatten()

# 그래프 그리기
plt.figure(figsize=(14, 10))
#plt.plot(new_dates[seq_len:seq_len+len(prediction2)], test_data[seq_len:seq_len+len(prediction2)], color='blue', label='Actual Level Price')
plt.plot(new_dates[seq_len:seq_len+len(prediction2)], prediction_1d, color='red', linestyle='--', label='Predicted Level Price')
plt.xlabel('Date')
plt.ylabel('Level Price(m)')
plt.title('Actual and Predicted Level Price')
plt.legend()
plt.show()

######################################################################

# 데이터에 없는 새로운 날짜 범위의 예측 수위 그래프 안의 위험 수위를 초과하는 예측 및 해당 날짜 표시
print("Start Date:", new_dates[0])
print("End Date:", new_dates[-1])

# 우이천 위험수위: 32.81
# 탄천 위험수위: 22.67

# 위험 수위를 초과하는 예측 및 해당 날짜 표시
threshold2 = 17.5 # 위험수위, 현제는 확인을 위한 임시 값
print(f"현제 설정된 위험수위 : {threshold2}")
dangerous_predictions = []
for i in range(len(prediction2)):
    if seq_len + i < len(test_dates):
        if prediction2[i] >= threshold2:
            dangerous_predictions.append((prediction2[i], test_dates[seq_len + i]))

if len(dangerous_predictions) > 0:
    print("위험 수위를 초과하는 예측:")
    for pred, date in dangerous_predictions:
        print("예측:", pred, "날짜:", date)
else:
    print("위험 수위를 초과하는 예측이 없습니다.")

######################################################################
```
